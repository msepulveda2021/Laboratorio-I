(d) codigo de graficos $(V_{\infty}-V)/V_{\infty}$ versus $t$.

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 5.5, 12)

V_i = 5.4
V = np.array([0.000, 2.718, 4.068, 4.739, 4.739, 5.237, 5.319, 5.360, 5.380, 5.390, 5.395, 5.398])

V_f = (V_i-V)/V_i


fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=1, ncols=4, figsize=(16,4), constrained_layout=True)

ax1.plot(t, V_f, marker='o', markersize=4)
ax1.set_ylabel('Voltaje $[V]$')
ax1.set_xlabel('Tiempo $[s]$')

ax2.plot(t, V_f, marker='o', markersize=4)
ax2.semilogx()
ax2.set_xlabel('Tiempo $[s]$')

ax3.plot(t, V_f, marker='o', markersize=4)
ax3.semilogy()
ax3.set_xlabel('Tiempo $[s]$')

ax4.plot(t, V_f, marker='o', markersize=4)
ax4.loglog()
ax4.set_xlabel('Tiempo $[s]$')

plt.show()
```

(e) codigo de grafico $\ln[(V_{\infty}-V)/V_{\infty}]$ versus $t$

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 5.5, 12)

V_i = 5.4
V = np.array([0.000, 2.718, 4.068, 4.739, 4.739, 5.237, 5.319, 5.360, 5.380, 5.390, 5.395, 5.398])

V_o = np.log((V_i-V)/V_i)

plt.plot(t, V_o, marker='o', markersize=4)
plt.xlabel('Tiempo $[s]$')
plt.ylabel('Voltaje $[V]$')
plt.show()
```
