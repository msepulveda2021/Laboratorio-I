(d) Código del gráficos $\left(V_{\infty}-V\right)/V_{\infty}$ versus $t$.

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 5.5, 12) #tiempo
t_err = 12*[0.25] #error de medicion del tiempo

V_i = 5.4 #Voltaje para tiempos muy grandes

V = np.array([0.000, 2.718, 4.068, 4.739, 4.739, 5.237, 5.319, 5.360, 5.380, 5.390, 5.395, 5.398]) #Voltajes medidos
V_err = 12*[0.0005] #error de medicion voltajes

V_f = (V_i-V)/V_i #Voltaje despues de aplicarle la funcion pedida
V_ferr = ((V+V_i)/(V_i**2))*0.0005 #propagacion de error

fig, (ax1, ax2, ax3, ax4) = plt.subplots(nrows=1, ncols=4, figsize=(16,4), constrained_layout=True)

ax1.errorbar(t, V_f, V_ferr, t_err, marker='o', markersize=4, capsize=4)
ax1.set_ylabel('Voltaje $[V]$')
ax1.set_xlabel('Tiempo $[s]$')

ax2.errorbar(t, V_f, V_ferr, t_err, marker='o', markersize=4, capsize=4)
ax2.semilogx()
ax2.set_xlabel('Tiempo $[s]$')

ax3.errorbar(t, V_f, V_ferr, t_err, marker='o', markersize=4, capsize=4)
ax3.semilogy()
ax3.set_xlabel('Tiempo $[s]$')

ax4.errorbar(t, V_f, V_ferr, t_err, marker='o', markersize=4, capsize=4)
ax4.loglog()
ax4.set_xlabel('Tiempo $[s]$')

plt.show()
```

Para el cálculo de la propagación del error, se hizo el siguiente procedimiento,

Sea $~f(V, V_{\infty}) = \frac{V_{\infty}-V}{V_{\infty}}$

$$ err_{f} = \left|\frac{\partial f}{\partial V}\right| \cdot err_{V} + \left|\frac{\partial f}{\partial V_{\infty}}\right| \cdot err_{V_{\infty}} $$

$$ err_{f} = \frac{1}{V_{\infty}} \cdot err_{V} + \frac{1}{V_{\infty}^{2}} \cdot err_{V_{\infty}} $$

como $ err_{V} = err_{V_{\infty}} $

$$ err_{f} = \left( \frac{V+V_{\infty}}{V_{\infty}^{2}} \right) \cdot err_{V} $$



(e) Código del gráfico $\ln[(V_{\infty}-V)/V_{\infty}]$ versus $t$

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 5.5, 12)
t_err = 12*[0.25]

V_i = 5.4

V = np.array([0.000, 2.718, 4.068, 4.739, 4.739, 5.237, 5.319, 5.360, 5.380, 5.390, 5.395, 5.398])
V_err = 12*[0.0005]

V_o = np.log((V_i-V)/V_i)
V_oerr = ((V_i+V)/(((V_i)**2)-(V_i*V)))*0.0005


plt.figure(figsize=(8,7))
plt.errorbar(t, V_o,V_oerr, t_err, marker='o', markersize=4, capsize=4)
plt.xlabel('Tiempo $[s]$')
plt.ylabel('Voltaje $[V]$')
plt.show()
```

Del mismo modo que en (d), para la propagación del error, se hizo el siguiente procedimiento,

Sea $~f(V, V_{\infty}) = \ln{\left(\frac{V_{\infty}-V}{V_{\infty}}\right)}$

$$ err_{f} = \left|\frac{\partial f}{\partial V}\right| \cdot err_{V} + \left|\frac{\partial f}{\partial V_{\infty}}\right| \cdot err_{V_{\infty}} $$

$$ err_{f} = \left|\frac{V}{V_{\infty}^{2}-V \cdot V_{\infty}}\right| \cdot err_{V} + \left| \frac{1}{V-V_{\infty}}\right| \cdot err_{V_{\infty}} $$

como $~err_{V}$ = $err_{V_{\infty}}$ y $~V_{\infty} > V$, tenemos que, 

$$ err_{f} = \left( \frac{V}{V_{\infty}^{2}-V \cdot V_{\infty}} + \frac{V_{\infty}}{V_{\infty}^{2}-V \cdot V_{\infty}} \right) \cdot err_{V} $$

$$ err_{f} = \left( \frac{V_{\infty}+V}{V_{\infty}^{2}-V \cdot V_{\infty}} \right) \cdot err_{V} $$
